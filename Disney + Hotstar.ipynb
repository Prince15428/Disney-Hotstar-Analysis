# Introduction and Data Loading

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style="whitegrid")
plt.style.use("ggplot")

# Load the dataset
data = pd.read_csv("C:\\Users\\Prince Das\\Downloads\\archive\\disney_plus_shows.csv")

# Display the first 5 rows to understand the structure
data.head()
```

# Data Exploration

```python
# Checking basic info of dataset
data.info()

# Checking for missing values
missing_values = data.isnull().sum()
print("Missing values per column : ")
print(missing_values[missing_values > 0])

# Statistical summary of the dataset
data.describe(include = "all")
```

# Data Cleaning

```python
# Drop columns with excessive missing values
#columns_to_drop = ["metascore" , "awards"]
#data.drop(columns= columns_to_drop, inplace = True)

#Filling missing values for "type" and "runtime"
#data["type"].fillna("movie", inplace=True)

# Convert the 'runtime' column to string type to use .str methods
data["runtime"] = data["runtime"].astype(str)

# Extract numeric part from 'runtime' and convert to float
data["runtime"] = data["runtime"].str.extract(r"(\d+)").astype(float)

#Fill missing values in "runtime" with the median value
median_runtime = data["runtime"].median() 
data["runtime"] = data["runtime"].fillna(median_runtime)

# Fill missing categorical data with 'Unknown'
for col in ['rated', 'genre', 'director', 'writer', 'actors', 'language', 'country']:
    data[col] = data[col].fillna('Unknown')

# Fill missing "imdb_rating" with the median value
median_imdb_rating = data["imdb_rating"].median()
data["imdb_rating"] = data["imdb_rating"].fillna(median_imdb_rating)

# Convert "imdb_votes" to numeric, handle commas, fill missing values
# Ensure all values in 'imdb_votes' are treated as strings
data['imdb_votes'] = data['imdb_votes'].astype(str)

# Replace commas with an empty string to remove them
data['imdb_votes'] = data['imdb_votes'].str.replace(',', '')

# Convert to numeric, coercing errors to NaN
data['imdb_votes'] = pd.to_numeric(data['imdb_votes'], errors='coerce')

# Calculate the median of the 'imdb_votes' column
median_imdb_votes = data['imdb_votes'].median()

# Fill missing values with the median value
data['imdb_votes'] = data['imdb_votes'].fillna(median_imdb_votes)

# Display the cleaned dataset summary
data.info()
```
